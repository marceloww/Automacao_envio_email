```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1
# Passo 1: Entrar no sistema (no nosso caso vai ser entrar no link)
pyautogui.hotkey("ctrl" , "t")
pyperclip.copy("https://drive.google.com/drive/folders/11yDwxgplfldJyPM4WQiNbujYKQEAU5ry")
pyautogui.hotkey("ctrl" , "v")
pyautogui.press("enter")

#esperar o site carregar
time.sleep(5)

# Passo 2: Navegar no sistema e encontrar a base de dados (entrar na pasta exportar)
pyautogui.click(x=368, y=285, clicks=2)
time.sleep(2)

# Passo 3: Download da base de dados
pyautogui.click(button="right",x=341, y=289)
pyautogui.click(x=541, y=747)
```

```python
# Passo 4: Calcular os indicadores (faturamento, quantidade de produtos)
import pandas as pd

tabela = pd.read_excel(r"C:\Users\Marcelo\Downloads\Vendas - Dez.xlsx")
display(tabela)
```

```python
quantidade = tabela["Quantidade"].sum() #soma da coluna quantidade

faturamento = tabela["Valor Final"].sum() #soma da coluna valor final

print(quantidade)
print(faturamento)
```

```python
# Passo 5: Entrar no E-mail
pyautogui.hotkey("ctrl" , "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl" , "v")
pyautogui.press("enter")
time.sleep(4)

# Passo 6: Mandar um email para a diretoria com os indicadores

#Clicar no botão escrever
pyautogui.click(x=72, y=203)
time.sleep(1)
#escrever o destinatátio (pra quem eu vou mandar)
pyautogui.write("thaisaaj@hotmail.com.br")
pyautogui.press("Tab") #seleciona o email
pyautogui.press("Tab") #muda para o campo assunto
#escrever o assunto
pyautogui.write("Relatorio de vendas")
pyautogui.press("Tab") #muda para o campo corpo email
#escrever o corpo do email
texto = f"""
Prezados, bom dia!

Segue o faturamento obtido ontem no valor de R$ {faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Att.,
Marcelo Faria Macedo Goncalves
"""

pyautogui.write(texto)


#enviar o email
pyautogui.hotkey("ctrl" , "enter")
```
